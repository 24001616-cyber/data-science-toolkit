4. Create a random numpy array of size 6x3 and transpose it

```python
c = np.random.randint(0, 10, size=(6, 3))
print("Original array (6x3):")
print(c)
```

```python
print("\nTransposed array (3x6):")
print(c.T)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
n, p, m = 4, 5, 3   # example sizes

# Random arrays
d = np.random.randint(0, 10, size=(n, p))   # shape (n, p)
e = np.random.randint(0, 10, size=(p, m))   # shape (p, m)

print("Array d (n x p):")
print(d)
print("\nArray e (p x m):")
print(e)
```

```python
f = np.dot(d, e)   # or d @ e
print("\nDot product result (n x m):")
print(f)
```
